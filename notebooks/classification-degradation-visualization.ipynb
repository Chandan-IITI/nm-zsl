```python
import matplotlib.pyplot as plt
# import seaborn as sns; sns.set()
import numpy as np

data = {
    'CUB (pretrained model)': {
        'scales': [64, 128, 200, 256, 448],
        'accuracies': [27, 55, 69, 73.5, 78],
    },
    'AwA2': {
        'scales': [32, 64, 128, 256],
        'accuracies': [36, 48, 63, 77.5]
    },
#     'Tiny ImageNet': {
#         'scales': [16, 32, 64],
#         'accuracies': [20, 28.5, 39.5]
#     }
}

all_scales = np.unique([s for d in data for s in data[d]['scales']])
all_scales = sorted(all_scales, reverse=True)

plt.title('Classification degradation for ResNet-18')

for dataset in data:
    plt.plot([all_scales.index(s) for s in data[dataset]['scales']], data[dataset]['accuracies'],
             label=dataset, marker='.', markersize=10)
    
# plt.grid(linewidth=0.5, color='222')
plt.grid()
plt.xlabel('Image resolution')
plt.ylabel('Final test accuracy')
plt.xticks(np.arange(len(all_scales)), all_scales)
plt.legend()
plt.savefig('/Users/universome/Downloads/classification-degradation.pdf')
```
